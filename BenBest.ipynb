```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
players1 = pd.read_csv("phs_2020_1.csv").drop(['start_time'], axis=1)
players2 = pd.read_csv("phs_2020_2.csv").drop(['start_time'], axis=1)
combine = [players1, players2]
data = pd.concat(combine)
#np.where(df.stat_name == 'All Damage Done')[0]
#df[df['stat_name'] == 'All Damage Done' and df['hero_name'] == 'Mercy']
data.drop(data.index[data['tournament_title'] != 'OWL 2020 Regular Season'], inplace = True)
data.drop(['tournament_title'], axis = 1, inplace = True)
```

```python
cols = ['esports_match_id', 'map_type', 'map_name', 'player_name', 'team_name', 'hero_name', 'stat_name', 'stat_amount']
data = data.reindex(columns = cols)
data = data[data["hero_name"] != "All Heroes"]
stats = ['Hero Damage Done', 'Assists', 'Barrier Damage Done', 'Deaths', 'Eliminations', 'Objective Kills', 'Objective Time', 'Ultimates Earned - Fractional', 'Final Blows', 'Damage Blocked', 'Healing Done', 'All Damage Done']
data = data.loc[data["stat_name"].isin(stats)]
data['stat_amount'] = data['stat_amount'].round(decimals = 2)
```

```python
test = data.groupby(['esports_match_id', 'map_type', 'map_name', 'team_name', 'player_name','hero_name']).agg(list)
#tab = test[['hero_name', 'stat_name', 'stat_amount']].copy().to_frame()
test = data[['stat_name', 'stat_amount', 'player_name','esports_match_id','hero_name','map_name', 'team_name']].copy().reset_index()
test = test[['stat_name', 'stat_amount', 'player_name','esports_match_id','hero_name','map_name', 'team_name']]
test = test.pivot_table(values='stat_amount', index=[test['esports_match_id'],test['team_name'],  test['map_name'], test['player_name'], test['hero_name']], columns='stat_name', aggfunc='first')
test = test.reset_index()
test = test.fillna(0)
test['map_name'].replace({'King\'s Row': 'Kings Row'}, inplace = True)
test
```

```python
maps = pd.read_csv("match_map_stats.csv")
maps['map_name'].replace({'King\'s Row': 'Kings Row'}, inplace = True)
```

```python
cleaned_maps = pd.DataFrame()
maps.query("`match_id` >= 30991", inplace = True)
maps.query("`match_id` <= 36231", inplace = True)
cleaned_maps['match_id'] = maps['match_id']
cleaned_maps['team_one'] = maps['team_one_name']
cleaned_maps['team_two'] = maps['team_two_name']
cleaned_maps['map_winner'] = maps['map_winner']
cleaned_maps['map_name'] = maps['map_name']
cleaned_maps['game_number'] = maps['game_number']
cleaned_maps['match_winner'] = maps['match_winner']
cleaned_maps.drop_duplicates(inplace = True)
cleaned_maps
```

```python
new_frame = pd.DataFrame(columns = cleaned_maps.columns)
for index, row in cleaned_maps.iterrows():
    if row['team_two'] < row['team_one']:
        new_frame.loc[index] = [row['match_id'], row['team_two'], row['team_one'], row['match_winner'], row['map_name'], row['game_number'], row['map_winner']]
    else:
        new_frame.loc[index] = [row['match_id'], row['team_one'], row['team_two'], row['match_winner'], row['map_name'], row['game_number'], row['map_winner']]
cleaned_maps = new_frame
cleaned_maps      
```

```python
map_winner = []
match_winner = []
game_number = []
temp1 = pd.DataFrame()
for index,row in test.iterrows():
    map = str(row['map_name'])
    match_id = int(row['esports_match_id'])
    #print(map)
    #print(match_id)
    temp = cleaned_maps[cleaned_maps['match_id'] == match_id]
    temp1 = temp[temp['map_name'] == map]
    #print(temp1)
    
    #print('-----')
    map_winner.append(temp1.iloc[0]['map_winner'])
    game_number.append(temp1.iloc[0]['game_number'])
    match_winner.append(temp1.iloc[0]['match_winner'])
test1 = test
test1['map_winner'] = map_winner
test1['match_winner'] = match_winner
test1['game_number'] = game_number
test1 = test1[['esports_match_id', 'team_name', 'map_name', 'game_number', 'map_winner','match_winner','player_name','hero_name','Assists','Barrier Damage Done', 'Deaths','Eliminations','Healing Done','Hero Damage Done','Objective Kills', 'Objective Time']]
test1
```

```python
hero_df = test.drop(['match_winner'], axis = 1)
winners = []
for index,row in hero_df.iterrows():
    if (row['team_name'] != row['map_winner']):
        winners.append("L")
    else:
        winners.append("W")
hero_df['map_winner'] = winners
hero_df
```

```python
map_names = pd.unique(hero_df['map_name'])
hero_names = pd.unique(hero_df['hero_name'])
```

```python
hero_df['map_winner'] = hero_df['map_winner'].replace('W', '1')
hero_df['map_winner'] = hero_df['map_winner'].replace('L', '0')
hero_df['map_winner'] = hero_df['map_winner'].astype('int')
heroes = []
winners = []
maps = []
for map in map_names:
    for hero in hero_names:
        temp_hero = hero_df.loc[hero_df['hero_name'] == hero] 
        scores = temp_hero.loc[temp_hero['map_name'] == map]['map_winner']
        if (scores.size) == 0:
            winners.append(0)
        else:
            winners.append((sum(scores) / scores.size) * 100)
        heroes.append(hero)
        maps.append(map)
win_pct = pd.DataFrame()
win_pct['Hero'] = heroes
win_pct['Map'] = maps
win_pct['Win_Pct'] = winners
```

```python
win_pct['Win_Pct'] = win_pct['Win_Pct'].round(decimals = 2)
win_pct
```

```python
damagers = ['Ashe', 'Bastion', 'McCree', 'Echo', 'Genji', 'Hanzo', 'Junkrat', 'Mei', 'Pharah', 'Reaper', 'Soldier: 76', 'Sombra', 'Symmetra', 'Torbjörn', 'Tracer', 'Widowmaker']
tanks = ['D.Va', 'Doomfist', 'Orisa', 'Reinhardt', 'Roadhog', 'Sigma', 'Winston', 'Wrecking Ball', 'Zarya']
supports = ['Ana', 'Baptiste', 'Brigitte', 'Lúcio', 'Mercy', 'Moira', 'Zenyatta']
hero_class = []
classes = ['Damage','Support', 'Tank']
for index,row in win_pct.iterrows():
    if row['Hero'] in damagers:
        hero_class.append('Damage')
    if row['Hero'] in tanks:
        hero_class.append('Tank')
    if row['Hero'] in supports:
        hero_class.append('Support')
win_pct['Class'] = hero_class
top_three_df = pd.DataFrame()
for map in map_names:
    temp_df = win_pct.loc[win_pct['Map'] == map]
    for class_type in classes:
        temp_df = win_pct.loc[win_pct['Map'] == map]
        temp_df = temp_df.loc[win_pct['Class'] == class_type]
        temp_df = temp_df.sort_values(by=['Win_Pct'], ascending=False).head(2)
        combine = [top_three_df, temp_df]
        top_three_df = pd.concat(combine)
top_three_df
```

```python
win_pct = win_pct.sort_values(by=['Map', 'Hero'], ascending = True)
win_pct
```

```python
average_table = test
#average_table.drop('game_number', axis = 1, inplace=True)
#average_table.drop('map_winner', axis = 1, inplace=True)
average_table = average_table.groupby(['esports_match_id', 'map_name', 'hero_name']).mean().reset_index()
average_table
```

```python
classes = [damagers, supports, tanks]
class_names = ["Damage", "Support", "Tank"]
newest_df = pd.DataFrame()
for curr_class, class_name in zip(classes, class_names):
    # Take rows of only heroes of the current class
    curr_table = average_table.loc[average_table['hero_name'].isin(curr_class)].copy()
    # Define equations we will use to calculate overall efficiency and create efficiency column
    def healer_efficiency(row):
        if (class_name == "Damage"):
            return (2*row['Assists'] + 2*row['Barrier Damage Done'] - 1.5*row['Deaths'] + 2.5*row['Eliminations'] + 2*row['All Damage Done']
                    + 2*row['Final Blows'] + row['Healing Done'] + 2*row['Objective Kills'] + row['Objective Time'] + 1.5*row['Ultimates Earned - Fractional'])
        elif (class_name == "Support"):
            return (2*row['Assists'] + 1.5*row['Barrier Damage Done'] - 2*row['Deaths'] + 2*row['Eliminations'] + row['All Damage Done']
                    + row['Final Blows'] + 2.5*row['Healing Done'] + 1.5*row['Objective Kills'] + 1.5*row['Objective Time'] + 1.5*row['Ultimates Earned - Fractional'])
        else:
            return (1.5*row['Assists'] + row['Barrier Damage Done'] + 1.5*row['Damage Blocked'] - 3*row['Deaths'] + 2*row['Eliminations'] + row['All Damage Done']
                    + row['Final Blows'] + row['Healing Done'] + 1.5*row['Objective Kills'] + 1.5*row['Objective Time'] + row['Ultimates Earned - Fractional'])
            
    curr_table['Efficiency'] = curr_table.apply(lambda row: healer_efficiency(row), axis=1)
    
    curr_table = curr_table.groupby(['hero_name', 'map_name']).mean().reset_index()
    curr_table['Efficiency'] = curr_table['Efficiency'].round(decimals = 2)
    

    newest_df = pd.concat([newest_df, curr_table])
```

```python
newest_df = newest_df.sort_values(by=['map_name', 'hero_name'], ascending = True)
newest_df
```

```python
maps = []
heroes = []
efficiencies = []
for map in map_names:
    for hero in hero_names:
        temp_hero = newest_df.loc[newest_df['hero_name'] == hero] 
        score = temp_hero.loc[temp_hero['map_name'] == map]['Efficiency'].sum()
        maps.append(map)
        heroes.append(hero)
        efficiencies.append(score)
effic_df = pd.DataFrame()
effic_df['Hero'] = heroes
effic_df['Map'] = maps
effic_df['Efficiency'] = efficiencies
effic_df = effic_df.sort_values(by=['Map', 'Hero'], ascending = True)
```

```python
win_pct['Efficiency'] = effic_df['Efficiency']
win_pct['Master Key'] = (2*win_pct['Efficiency'] / (win_pct['Efficiency'].mean())) + (win_pct['Win_Pct'] / (win_pct['Win_Pct'].mean()))
win_pct = win_pct[['Hero','Map','Class','Win_Pct','Efficiency','Master Key']]
win_pct.sort_values(by=['Master Key'], ascending = False)
```

```python
damage_df = win_pct.loc[win_pct['Class'] == 'Damage'].sort_values(by=['Master Key'], ascending=False)
support_df = win_pct.loc[win_pct['Class'] == 'Support'].sort_values(by=['Master Key'], ascending=False)
tank_df = win_pct.loc[win_pct['Class'] == 'Tank'].sort_values(by=['Master Key'], ascending=False)
maps = []
damage_one = []
damage_two = []
support_one = []
support_two = []
tank_one = []
tank_two = []
for map in map_names:
    best_damage = damage_df.loc[damage_df['Map'] == map].head(2)
    best_damage_arr = np.array(best_damage['Hero'])
    damage_one.append(best_damage_arr[0])
    damage_two.append(best_damage_arr[1])
    best_support = support_df.loc[support_df['Map'] == map].head(2)
    best_support_arr = np.array(best_support['Hero'])
    support_one.append(best_support_arr[0])
    support_two.append(best_support_arr[1])
    best_tank = tank_df.loc[tank_df['Map'] == map].head(2)
    best_tank_arr = np.array(best_tank['Hero'])
    tank_one.append(best_tank_arr[0])
    tank_two.append(best_tank_arr[1])
    maps.append(map)
final_team_composition = pd.DataFrame()
final_team_composition['Map'] = maps
final_team_composition['Damage #1'] = damage_one
final_team_composition['Damage #2'] = damage_two
final_team_composition['Tank #1'] = tank_one
final_team_composition['Tank #2'] = tank_two
final_team_composition['Support #1'] = support_one
final_team_composition['Support #2'] = support_two
```
